```python
print(new_data.head())
```

```python
def image_exists(image_path):
    return os.path.isfile(image_path)
```

```python
df_cleaned = new_data[new_data['Image_path'].apply(image_exists)]
```

```python
print(df_cleaned)
```

```python
# Define the old part and new part of the image path
old_path_part = "train_images"
new_path_part = "retrain_images"

# Replace the old path part with the new one
df_cleaned['Image_path'] = df_cleaned['Image_path'].str.replace(old_path_part, new_path_part, regex=False)

print(df_cleaned)
```

```python
df_cleaned=df_cleaned.reset_index(drop=True)
df_cleaned
```

```python
df_cleaned.to_csv(r"C:\Users\varsh\OneDrive\Desktop\notebook\Fast_Furious_Insured\data\retrain_data")
```
